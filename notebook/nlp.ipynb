## Preparing data

```python
from pathlib import Path
dataset_path_pos=Path("../data/train/pos")
dataset_path_neg=Path("../data/train/neg")
```

```python
pos_files = list(dataset_path_pos.glob("*.txt"))
neg_files = list(dataset_path_neg.glob("*.txt"))


print(len(pos_files))
```

```python
sample_file=pos_files[0]
text=sample_file.read_text(encoding='utf-8')
print(text[:100])
```

```python
# Reading all the Review
def read_review(files): 
    reviews=[]
    for file in files:
        text=file.read_text(encoding='utf-8')
        reviews.append(text)
    return reviews
```

```python
pos_review=read_review(pos_files)
```

```python
neg_review=read_review(neg_files)
```

```python
def store_data(reviews,rev_type):
    data=[]
    for review in reviews:
        data.append({"review": review,
                    "sentiment":rev_type})
    return data
```

```python
pos_data=store_data(pos_review,"positive")
neg_data=store_data(neg_review,"negative")
```

```python
all_data=pos_data+neg_data
```

```python
import pandas as pd 
df= pd.DataFrame(all_data, index=True)
```

```python
df=df.sample(frac=1,random_state=42).reset_index(drop=True)
```
