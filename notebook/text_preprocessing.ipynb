### HTML Removal 

```python
import regex as re
```

```python
def html_remover(text):
    if isinstance(text,str):
        return re.sub(r"<.*?>","",text)
    return text
```

```python
df['review']=df['review'].apply(html_remover)
```

```python
df['review'][0]
```

```python
import string
def punc_clean(text):
    pun=string.punctuation
    samp_tex=str()
    for i in text:
        if i not in pun:
            samp_tex += i 
    return samp_tex
```

```python
df['review']=df['review'].apply(punc_clean)
```

```python
df['review'][0]
```

## Tokenization

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')
nltk.download('punkt')
```

```python
def toke(text):
    tokens=word_tokenize(text)
    return tokens
```

```python
df['clean_review']=df['review'].apply(toke)
```

```python
df = df[['review', 'clean_review', 'sentiment']]
```

```python
df.sample(5)
```

## Stop words

```python
from nltk.corpus import stopwords
```

```python
def stop_word(text):
    stop_words=set(stopwords.words('english'))
    filtered=[
        word for word in text
        if word.lower() not in stop_words
    ]
    return filtered
```

```python
df['clean_review']=df['clean_review'].apply(stop_word)
```
