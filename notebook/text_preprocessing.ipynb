# Text preprocessing

### HTML Removal 

```python
import regex as re
```

```python
def html_remover(text):
    if isinstance(text,str):
        return re.sub(r"<.*?>","",text)
    return text
```

```python
df['review']=df['review'].apply(html_remover)
```

```python
df['review'][0]
```

```python
import string
def punc_clean(text):
    pun=string.punctuation
    samp_tex=str()
    for i in text:
        if i not in pun:
            samp_tex += i 
    return samp_tex
```

```python
df['review']=df['review'].apply(punc_clean)
```

## Tokenization

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')
nltk.download('punkt')
```

```python
def toke(text):
    tokens=word_tokenize(text)
    return tokens
```

```python
df['clean_review']=df['review'].apply(toke)
```

```python
df = df[['review', 'clean_review', 'sentiment']]
```

```python
df.sample(5)
```

## Stop words

```python
from nltk.corpus import stopwords
```

```python
stop_words=set(stopwords.words('english'))
stop_words.remove('not')
stop_words.remove('no')
stop_words.remove('nor')
```

```python
def stop_word(text):
    # stop_words=set(stopwords.words('english'))
    filtered=[
        word for word in text
        if word.lower() not in stop_words
    ]
    return filtered
```

```python
df['clean_review']=df['clean_review'].apply(stop_word)
```

```python
df['clean_review']
```

## Lemmatization


```python
from nltk.stem import WordNetLemmatizer
def lemmatization(lis):
    lemmatizer=WordNetLemmatizer()
    l=list()
    for word in lis:
        a=lemmatizer.lemmatize(word)
        l.append(a)
    return l
```

```python
df['clean_review']=df['clean_review'].apply(lemmatization)  
```

```python
df['clean_review'][0]
```

```python
df.head()
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df['clean_review']=df['clean_review'].apply(lambda x:" ".join(x))
```

```python
df['clean_review']
```

```python
tfidf=TfidfVectorizer(max_features=5000)

X=tfidf.fit_transform(df['clean_review'])
```

```python
y=df['sentiment']
```

## Train and Test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42, test_size=0.1)
```

```python
X_train.shape
```

## Training Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model=LogisticRegression(max_iter=1000)
```

```python
lr_model.fit(X_train,y_train)
```

```python
y_pred=lr_model.predict(X_test)
```
